### Introduction to matplotlib

```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.plot();
```

```python
x=[2,13,23,43,54]
y=[13,35,56,59,65]

fig,ax= plt.subplots(figsize=(5,5))

ax.plot(x,y)

ax.set(title="Sample_graph",
      xlabel="x-axis",
      ylabel="y-axis")

fig.savefig("./images/sample-plot.png")
```

```python
x=np.linspace(1,10,50)
fig,ax=plt.subplots(figsize=(7,7))
ax.plot(x,x**3)
```

```python
fig,ax=plt.subplots()
ax.scatter(x,np.cos(x));
ax.scatter(x,np.sin(x))
```

```python
x2=np.random.randn(100)
fig,ax=plt.subplots()

ax.hist(x2);
```

```python
fig,((ax1,ax2),(ax3,ax4))=plt.subplots(nrows=2,ncols=2,figsize=(20,10))

ax1.plot(x,x*2);


ax2.bar(x[:10],x[:10]**3);

ax3.barh(x,x/2);

ax4.scatter(x,x*x);
```

```python
car_sales=pd.read_csv("export_demo.csv")
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str.replace('[\$\,\.]','')
```

```python
car_sales
```

```python
car_sales["Price"]=car_sales["Price"].str[:-2]
```

```python
car_sales
```

```python
car_sales["Date"]=pd.date_range("01/01/2000",periods=len(car_sales))
```

```python
car_sales
```

```python
car_sales["Total Sale"]=car_sales["Price"].astype(int).cumsum()
```

```python
car_sales
```

```python
car_sales.plot(x="Date",y="Total Sale")
```
