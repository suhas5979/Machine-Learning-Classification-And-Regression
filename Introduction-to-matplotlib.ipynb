```python
heart_disease= pd.read_csv("heart-disease.csv")
```

```python
heart_disease
```

```python
heart_disease.head(20).plot.bar(subplots=True,figsize=(20,50));
```

```python
over_60=heart_disease[heart_disease["age"]>60]
```

```python
over_60
```

```python
fig ,ax=plt.subplots(figsize=(10,6))
over_60.plot(x='age',y='chol',kind='scatter',ax=ax,c='target')
#ax.set_ylim([0,2000])
```

```python
fig,ax=plt.subplots(figsize=(10,6))

scatter= ax.scatter(x=over_60["age"],y=over_60["chol"],c=over_60["target"])
ax.set(title="Heart-Disease",xlabel="Age",ylabel="Choloestrol")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(y=over_60["chol"].mean(),linestyle='--')
```

##### Styling


```python
plt.style.use('seaborn-whitegrid')
fig,ax=plt.subplots(figsize=(10,6))

scatter= ax.scatter(x=over_60["age"],y=over_60["chol"],c=over_60["target"],cmap='cool')
ax.set(title="Heart-Disease",xlabel="Age",ylabel="Choloestrol")
ax.legend(*scatter.legend_elements(),title="Target");
ax.axhline(y=over_60["chol"].mean(),linestyle='--')
```

```python
fig ,(ax0,ax1)=plt.subplots(nrows=2,ncols=1,figsize=(15,12),sharex=True)
scatter=ax0.scatter(x=over_60["age"],y=over_60["chol"],c=over_60["target"],cmap='cool')
ax0.set(title="Heart-Disease",xlabel="Age",ylabel="Choloestrol")
ax0.legend(*scatter.legend_elements(),title="Target");
ax0.axhline(y=over_60["chol"].mean(),linestyle='--')
scatter=ax1.scatter(x=over_60["age"],y=over_60["chol"],c=over_60["target"],cmap='brg')
ax1.set(title="Heart-Disease",xlabel="Age",ylabel="Choloestrol")
ax1.legend(*scatter.legend_elements(),title="Target");
ax1.axhline(y=over_60["chol"].mean(),linestyle='--')
```
